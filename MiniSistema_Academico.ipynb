```
# Clase Estudiante
class Estudiante:
    def __init__(self, documento, nombre):
        self.documento = documento
        self.nombre = nombre
        self.cursos = []

    def agregar_curso(self, curso):
        self.cursos.append(curso)

    def mostrar_cursos(self):
        for curso in self.cursos:
            print(f"- {curso.nombre}")


# Clase Curso
class Curso:
    def __init__(self, nombre, intensidad):
        self.nombre = nombre
        self.intensidad = intensidad
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)

    def mostrar_estudiantes(self):
        print(f"Estudiantes en el curso {self.nombre}:")
        for est in self.estudiantes:
            print(f"- {est.nombre} ({est.documento})")


# Crear lista de cursos disponibles
lista_cursos = [
    Curso("Python", 40),
    Curso("JavaScript", 30),
    Curso("Base de Datos", 20)
]

# Crear lista de estudiantes
lista_estudiantes = [
    Estudiante("1001", "Ana Gómez"),
    Estudiante("1002", "Luis Torres"),
    Estudiante("1003", "María Pérez")
]

# Matricular estudiantes en cursos
lista_cursos[0].agregar_estudiante(lista_estudiantes[0])  # Ana en Python
lista_estudiantes[0].agregar_curso(lista_cursos[0])

lista_cursos[1].agregar_estudiante(lista_estudiantes[1])  # Luis en JavaScript
lista_estudiantes[1].agregar_curso(lista_cursos[1])

lista_cursos[0].agregar_estudiante(lista_estudiantes[2])  # María en Python
lista_estudiantes[2].agregar_curso(lista_cursos[0])
lista_cursos[2].agregar_estudiante(lista_estudiantes[2])  # María en BD
lista_estudiantes[2].agregar_curso(lista_cursos[2])

# Mostrar resultados
print("Listado de estudiantes por curso:\n")
for curso in lista_cursos:
    curso.mostrar_estudiantes()
    print()

print("Cursos de cada estudiante:\n")
for est in lista_estudiantes:
    print(f"{est.nombre} está matriculado en:")
    est.mostrar_cursos()
    print()
```
